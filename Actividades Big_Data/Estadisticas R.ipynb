1. Promedio de defunciones ocurridas a nivel nacional en el mes de su cumple durante el año 2021

```python
%%R
octubre.2021 <- data %>%
select(nombre, ends_with("10.2021")) %>%
filter(nombre == "Nacional" ) %>%
select(-nombre)

octubre.2021 %>%
t()%>%
mean()
```

 2. Número total de defunciones a nivel nacional en el año 2020

```python
%%R
año.2020 <- data %>%
select(nombre, ends_with("2020")) %>%
filter(nombre == "Nacional" ) %>%
select(-nombre)

año.2020 %>%
t()%>%
sum()
```

3. Estado con el mayor número de defunciones en el mes de Agosto 2020 (sin contar la Ciudad de México ni el Estado de México)

```python
%%R
data %>%
select(nombre, ends_with("08.2020")) %>%
filter(nombre != "DISTRITO FEDERAL", nombre != "MEXICO",nombre != "Nacional") %>%
mutate(num_def = rowSums(pick(where(is.numeric)))) %>%
arrange(desc(num_def))%>%
select(nombre,num_def)%>%
head(1)
```

4.    Mes con mayor número de defunciones a nivel nacional en el año 2022



```python
%%R
resultados <- data %>%
mutate(ene.2022 = rowSums(select(data,ends_with("01.2022"))),
       feb.2022 = rowSums(select(data,ends_with("01.2022"))),
       mar.2022 = rowSums(select(data,ends_with("01.2022"))),
       abr.2022 = rowSums(select(data,ends_with("01.2022"))),
       may.2022 = rowSums(select(data,ends_with("01.2022"))),
       jun.2022 = rowSums(select(data,ends_with("01.2022"))),
       jul.2022 = rowSums(select(data,ends_with("01.2022"))),
       ago.2022 = rowSums(select(data,ends_with("01.2022"))),
       sep.2022 = rowSums(select(data,ends_with("01.2022"))),
       oct.2022 = rowSums(select(data,ends_with("01.2022"))),
       nov.2022 = rowSums(select(data,ends_with("01.2022"))),
       dic.2022 = rowSums(select(data,ends_with("01.2022"))))%>%
filter(nombre == "Nacional")%>%
select(ene.2022,feb.2022,mar.2022,abr.2022,may.2022,jun.2022,jul.2022,ago.2022,sep.2022,oct.2022,nov.2022,dic.2022)
```

```python
%%R
resultados %>%
select(apply(resultados, 1, which.max))
```
